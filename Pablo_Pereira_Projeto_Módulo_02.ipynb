## Objetivo

O objetivo principal desta análise é verificar algumas hipóteses na tentativa de identificar os motivos da baixa adesão as campanhas nacionais de vacinação nos últimos anos. A baixa adesão culminou com surtos de Sarampo, que até então era considerada uma doença erradicada do Brasil desde [2016](https://www.paho.org/bra/index.php?option=com_content&view=article&id=5248:regiao-das-americas-e-declarada-livre-de-sarampo&Itemid=820), e que poderia indicar ainda que outras doenças como 
Poliomielite, que também já tinha sido erradicadas do Brasil em [1994](https://www.paho.org/bra/index.php?option=com_content&view=article&id=5735:folha-informativa-poliomielite&Itemid=820), possam retornar. 


## Resumo

Doenças como Sarampo e Poliomielite haviam sido erradicadas do território nacional através de ampla conscientização da população da importância das Campanhas Nacionais de Vacinação e, da adesão a mesma, mas os novos casos de Sarampo no [Brasil](https://www.paho.org/bra/index.php?option=com_content&view=article&id=5633:folha-informativa-sarampo&Itemid=1060) e no mundo alertam para o retorno de doenças possivelmente erradicadas diante de baixas taxas de cobertura vacinal.

Através das informações de bases de dados do [DataSUS - TabNet](http://tabnet.datasus.gov.br/cgi/tabcgi.exe?pni/cnv/cpniuf.def), realizei a análise da cobertura vacinal entre os anos 2015 à 2019 das seguintes vacinas:

1. **Poliomielite**
2. **Poliomielite(1º reforço)**
3. **Tríplice Viral** - Imunizante contra Sarampo, Caxumba e Rubéola
4. **Tríplice Viral - Segunda dose** - Imunizante contra Sarampo, Caxumba e Rubéola

De fato, a queda da adesão da população as Campanhas Nacionais de Imunização deixaram as metas de cobertural vacinal muito abaixo no ano de 2019, em destaque para o estado do Rio de Janeiro que não atingiu nem 50% da meta para os imunizantes analisados.

Verifiquei também queda de investimentos em publicidade do Ministério da Saúde para capanhas de utilidade pública em geral, incluindo vacinação. Somado a isso, houve um aumento de circulação de Fake News envolvendo vacinas nos meios digitais e, que foram confrontadas com termos buscados no google sobre o tema, onde também houve um aumento no ano de 2019.


## Problematização

A cobertura vacinal é um indicador que estima a proporção da população-alvo vacinada. Para o cálculo, utiliza-se o total de últimas doses do esquema da vacina de interesse, no numerador, dividido pela estimativa da população alvo, no denominador, multiplicado por 100. Para a população de menores de um ano e de um ano o denominador é extraído do **Sinasc (Sistema de Nascidos Vivos)**. A meta de CV é de 80% para a vacina contra o HPV, 90% para as vacinas BCG e Rotavírus e 95% para as demais vacinas.

Como amplamente divulgado na [mídia](https://g1.globo.com/bemestar/noticia/casos-de-sarampo-e-poliomelite-aumentaram-em-todo-o-mundo-diz-relatorio-da-oms.ghtml), a cobertura vacinal no Brasil vem sofrendo queda ano após ano, e os possíveis motivos são vários, como citados pela revista [Pesquisa FAPESP](https://revistapesquisa.fapesp.br/as-razoes-da-queda-na-vacinacao/) - Editada pela Fundação de Amparo à Pesquisa do Estado de São Paulo (FAPESP). 

Com isso, doenças que eram dadas como erradicadas do Brasil, como por exemplo sarampo, ressurgem com força no território nacional tendo inclusive como conseqüencia a perda do [certificado de país livre do vírus causador do Sarampo](https://oglobo.globo.com/sociedade/saude/surtos-de-sarampo-fazem-brasil-perder-certificado-de-pais-livre-do-virus-23534799) em 2019.

Nesta análise verificarei se a queda para a cobertura vacinal se confirma para os imunizantes **poliomielite e tríplice viral(Imunizante para Sarampo, Caxumba e Rubéola)**, entre os anos de 2015 e 2019, juntamente com a diminuição de investimentos em campanhas de conscientização, dado que muito desta baixa adesão se dá pela falta de informações. Abaixo citação da publicação da revista [Pesquisa FAPESP](https://revistapesquisa.fapesp.br/as-razoes-da-queda-na-vacinacao/) que levantou junto ao Ministério da Saúde, o que consideram possíveis causas para a baixa cobertura vacinal. 

>*O ministério aposta mais em cinco razões: a percepção enganosa dos pais de que não é preciso mais vacinar porque as doenças desapareceram; o desconhecimento de quais são os imunizantes que integram o calendário nacional de vacinação, todos de aplicação obrigatória; o medo de que as vacinas causem reações prejudiciais ao organismo; o receio de que o número elevado de imunizantes sobrecarregue o sistema imunológico; e a falta de tempo das pessoas para ir aos postos de saúde, que funcionam das 8h às 17h só nos dias úteis. Esses motivos são os mais mencionados nos monitoramentos feitos em municípios brasileiros em seguida às campanhas nacionais – no estado de São Paulo foram visitados 58 mil domicílios após a campanha de 2017. “Esses fatores podem interferir na cobertura vacinal, mas não são os únicos”, afirma Carla. “Também identificamos relatos de médicos e outros profissionais da saúde que aconselham as pessoas a não se vacinarem contra as doenças que não existem mais no país.”*

De todos os motivos apresentados acima pelo Ministério da saúde, percebemos o quão importante é o investimento em publicidade para suportar as Campanhas Nacionais de Vacinação. Todos os motivos apresentados são frutos de falta de informação, informação equivocada e falta de conscientização, que poderiam ser revertidos com investimentos em publicade bem estruturada.

## Glossário - Portal da Transparência

Glossário para os termos utilizados no Orçamento do Ministério da Saúde segundo o informado no Portal da Transperência.

**Valor Empenhado**
Valor resultante da fase de empenho, que é o primeiro estágio da despesa. Consiste na reserva de dotação orçamentária, ou seja, reserva de valores monetários autorizados para atender um fim específico. O empenho é registrado no momento da contratação do serviço, aquisição do material ou bem, obra ou amortização da dívida

**Valor Liquidado**
Valor resultante da fase de liquidação, segundo estágio da execução da despesa. Consiste na verificação do direito adquirido pelo credor, tendo por base os títulos e documentos comprobatórios do respectivo crédito. É nesse segundo estágio da execução da despesa que será cobrada a prestação dos serviços ou a entrega dos bens, ou ainda, a realização da obra. Envolve, portanto, todos os atos de verificação e conferência, desde a entrega do material ou a prestação do serviço até o reconhecimento da despesa

**Valor Pago**
Valor resultante da fase de pagamento, último estágio da execução da despesa. É o valor efetivamente entregue ao credor. Esse pagamento normalmente é efetuado por meio de crédito em conta bancária do favorecido

**Restos a Pagar**
Representam as despesas empenhadas e não pagas até 31 de dezembro, distinguindo-se as despesas processadas das não processadas. Entende-se por processadas e não processadas, respectivamente, as despesas liquidadas e as não liquidadas (art. 67, Decreto nr. 93.872, de 23 de dezembro de 1986). Fonte: Tesouro Nacional - http://manualsiafi.tesouro.fazenda.gov.br

O Portal apresenta não apenas os valores efetivamente pagos (gastos ou transferidos), mas também os valores empenhados e liquidados, a depender da etapa de execução da despesa a que se refiram. Para saber mais sobre a execução da despesa, acesse "Aprenda Mais/Entenda gestão pública/Execução da despesa pública".

**Deve-se referir ao glossário para entendimento dos valores apontados acima. Para a minha análise vou considerar os valores efetivamente pagos e os valores que restam a pagar mediante o que foi previsto no orçamento do ministério da sáude para publicidade.**

```
# Novo dataframe com o total de gastos realizadas e pendendentes para o Ministério da Saúde
publicity_health_expenses_total = publicity_health_expenses.groupby(publicity_health_expenses.index.year)[['Valor Pago', 'Valor Restos a Pagar Pagos']].sum()/1000000
publicity_health_expenses_total['Total'] = publicity_health_expenses_total[['Valor Pago', 'Valor Restos a Pagar Pagos']].sum(axis=1)
publicity_health_expenses_total
```

**Análise de Gastos Efetuados com publicidade pelo ministério da saúde nos anos de 2015 à 2019**

```
# Plotando o gráfico de gastos efetuados com publicidade
plt.figure(figsize=(12,8))

ax = sns.lineplot(data=publicity_health_expenses_total, markers=True, dashes=False)

ax.set_title('Despesas com Publicidade - Ministério da Saúde', fontsize=15)
ax.set_xlabel('Período analisado',fontsize=15)
ax.set_ylabel('Valor em milhões - (R$) ',fontsize=15)

ax.set_xticks(np.arange(2015,2020))
ax.set_yticks(np.arange(0,280,5), minor=True)
ax.set_yticks(np.arange(0,281,20))

plt.axhline(95, color='k', ls='--')

# Calcula o percentual de queda do Orçamento de Publicade entre 2018 e 2020
perc_loss = (publicity_health_expenses_total.loc[2019]['Valor Pago']*100)/publicity_health_expenses_total.loc[2018]['Valor Pago']

perc_loss = ((publicity_health_expenses_total.loc[2018]['Valor Pago']-publicity_health_expenses_total.loc[2019]['Valor Pago'])/publicity_health_expenses_total.loc[2018]['Valor Pago'])*100

ax.annotate(f'Queda de {perc_loss:.2f}% ', 
            xy=(2019,100),             
            xycoords='data',
            xytext=(-200, 0),
            fontsize=16,
            textcoords='offset points',
            arrowprops=dict(arrowstyle='->', color='black'),
            ha='center',
            va='center')

# Ajustando os tickers de limites de x
ax.set_xlim(2015,2019)
# Ajustando os tickers de limites de y
ax.set_ylim(0,280)
# Estilizando o Grid
ax.grid(which='major', linestyle='--')
ax.grid(which='minor', linestyle='dashed')
plt.show();
```

**Não há como verificar o quanto da verba orçamentária para publicidade do ministério da Saúde foi dedicada exclusivamente as campanhas de conscientização sobre vacinação, mas é possível verificar que o montante geral de gastos com publicidade foi drasticamente reduzido para o ano de 2019, com uma queda de 50,57% ano contra ano comparado com 2018**

```
# Novo dataframe com o total de gastos realizadas e pendendentes para o Ministério da Saúde
prev_publicity_health_expenses_total = publicity_health_expenses.groupby(publicity_health_expenses.index.year)[['Valor Empenhado','Valor Liquidado','Valor Pago', 'Valor Restos a Pagar Pagos']].sum()/1000000
prev_publicity_health_expenses_total['Ratio Empenhado vs Pago'] = ((prev_publicity_health_expenses_total['Valor Empenhado']-prev_publicity_health_expenses_total['Valor Pago'])/prev_publicity_health_expenses_total['Valor Pago'])*100

# Estilizando a visualização
cm = sns.light_palette("red", as_cmap=True)
table = prev_publicity_health_expenses_total[0:5].style.background_gradient(cmap=cm).set_precision(2)
table
```

**A hipótese da convergência de fatores como a diminuição do investimento em publicidade e campanhas de incentivo a vacinação pelo ministério da saúde, a disseminação das Fake News relacionadas as vacinas, responsabilizando as mesmas por casos de autismo ou seqüelas e culminando com o fortaleciamento do movimento antivacina, podem ter gerado a "tempestade perfeita" provocando a diminuição da cobertura vacinal em todos os estados do Brasil.** 

**Não há até o momento e, com os dados disponíveis, a possibilidade de analisar em detalhes os gastos com publicidade para a conscientização da importância das campanhas de imunização, pois os dados não estão nesta granularidade, apenas de maneira geral na rubrica "4641 - Publicidade de utilidade pública" onde avaliamos a sua redução em 50.57% contra o ano de 2018.** 

**Não foi possível também detectar diretamente o volume de Fake News atrelados as vacinas, mas o aumento considerável de buscas no google trends nos dá uma pista de que houve um aumento de dúvidas quando a efetividade e segurança dos imunizantes. Como pudemos observar nos dados do google Trends, houve um aumento de buscas por termos relacionados a campanhas antivacinas em diversos momentos. Por isso, a hipótese tende a ser verdadeira nesta análise, mas precisaríamos de mais dados para afirmar a mesma com maior grau de confiabilidade.**
